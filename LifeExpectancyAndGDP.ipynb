```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Reading data from the CSV file
df = pd.read_csv('all_data.csv')

# Converting necessary columns to appropriate data types
df['GDP'] = df['GDP'].astype(float)
df['Life expectancy at birth (years)'] = df['Life expectancy at birth (years)'].astype(float)
df['Year'] = df['Year'].astype(int)

print(df.head())
```

```python
# Plotting the trend of life expectancy over time for each of the six nations
plt.figure(figsize=(12, 6))
for country in df['Country'].unique():
    plt.plot(df[df['Country'] == country]['Year'], df[df['Country'] == country]['Life expectancy at birth (years)'], label=country)
plt.title('Trend of Life Expectancy Over Time for Six Nations')
plt.xlabel('Year')
plt.ylabel('Life Expectancy at Birth (years)')
plt.legend()
plt.show()
```

```python
# Plotting the trend of GDP over time for each of the six nations
# Plotting the trend of GDP over time for each of the six nations
plt.figure(figsize=(12, 6))
for country in df['Country'].unique():
    plt.plot(df[df['Country'] == country]['Year'], df[df['Country'] == country]['GDP'] / 1e9, label=country)  # dividing by 1e9 to convert to billions
plt.title('Trend of GDP Over Time for Six Nations')
plt.xlabel('Year')
plt.ylabel('GDP (in billions of dollars)')
plt.legend()
plt.show()
```

```python
# Checking the correlation between GDP and life expectancy
plt.figure(figsize=(10, 6))
sns.scatterplot(x='GDP', y='Life expectancy at birth (years)', data=df, hue='Country')
plt.title('Correlation between GDP and Life Expectancy')
plt.show()
```

```python
# Calculating the average life expectancy for each country
average_life_expectancy = df.groupby('Country')['Life expectancy at birth (years)'].mean()
print("Average life expectancy for each country:")
print(average_life_expectancy)
```

```python
# Visualizing the distribution of life expectancy across the six nations
plt.figure(figsize=(10, 5))
sns.histplot(df, x='Life expectancy at birth (years)', hue='Country', multiple='stack', bins=15, kde=True)
plt.title('Distribution of Life Expectancy Across Six Nations')
plt.xlabel('Life Expectancy at Birth (years)')
plt.ylabel('Frequency')
plt.show()
```
